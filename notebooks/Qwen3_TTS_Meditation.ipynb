## 3️⃣ Load model

Run this once per session. Choose your model variant below.

```python
import torch
import soundfile as sf
from IPython.display import Audio, display
from qwen_tts import Qwen3TTSModel

# ── Choose your model ──────────────────────────────────────────────────────
# Options:
#   "Qwen/Qwen3-TTS-12Hz-1.7B-CustomVoice"   ← preset voices + instruction control
#   "Qwen/Qwen3-TTS-12Hz-1.7B-VoiceDesign"   ← describe any voice freely
#   "Qwen/Qwen3-TTS-12Hz-1.7B-Base"          ← voice cloning
#   "Qwen/Qwen3-TTS-12Hz-0.6B-CustomVoice"   ← lighter, faster
MODEL_ID = "Qwen/Qwen3-TTS-12Hz-1.7B-CustomVoice"
# ──────────────────────────────────────────────────────────────────────────

print(f'Loading {MODEL_ID} ...')
model = Qwen3TTSModel.from_pretrained(
    MODEL_ID,
    device_map="cuda:0",
    dtype=torch.bfloat16,
    # attn_implementation="flash_attention_2",  # uncomment if flash-attn is installed
)
print('✅ Model loaded')

# Show available speakers (only relevant for CustomVoice models)
if hasattr(model, 'get_supported_speakers'):
    print('\nAvailable speakers:', model.get_supported_speakers())
    print('Available languages:', model.get_supported_languages())
```

## 4️⃣B — Generate with VoiceDesign (describe any voice)

Use this if you loaded `Qwen3-TTS-12Hz-1.7B-VoiceDesign`.

```python
# ── Configure ──────────────────────────────────────────────────────────────
SESSION_NAME = "sleep_meditation_designed"
LANGUAGE     = "English"
VOICE_DESC   = """
Soft, androgynous voice, mid-range, very slow and deliberate pace.
Slightly breathy, with a warm resonance. Like a whisper that fills the room.
Each word is spaced with intention. Calm to the point of near-stillness.
"""
TEXT = """
Let your eyes grow heavy. Let your thoughts drift like clouds across a quiet sky.
There is nothing you need to hold onto right now.
With each exhale, release a little more. Release the day. Release the effort.
You are floating gently toward rest.
"""
# ──────────────────────────────────────────────────────────────────────────

print('Generating audio with custom voice design...')
wavs, sr = model.generate_voice_design(
    text=TEXT.strip(),
    language=LANGUAGE,
    instruct=VOICE_DESC.strip(),
)

output_path = f"{OUTPUT_DIR}/{SESSION_NAME}.wav"
sf.write(output_path, wavs[0], sr)

print(f'✅ Saved to: {output_path}')
display(Audio(wavs[0], rate=sr))
```

## 4️⃣C — Voice Clone from reference audio

Use this if you loaded `Qwen3-TTS-12Hz-1.7B-Base`. Provide a ~3s reference clip.

```python
# ── Configure ──────────────────────────────────────────────────────────────
SESSION_NAME = "clone_meditation"
LANGUAGE     = "English"

# Reference audio: local path, URL, or (numpy_array, sample_rate) tuple
REF_AUDIO = "https://qianwen-res.oss-cn-beijing.aliyuncs.com/Qwen3-TTS-Repo/clone.wav"
REF_TEXT  = "Okay. Yeah. I resent you. I love you. I respect you. But you know what? You blew it!"

TEXT = """
Gently bring your attention to the space between your thoughts.
Not forcing, not grasping. Simply observing, with kindness.
You are the awareness behind the thoughts. Still. Spacious. Present.
"""
# ──────────────────────────────────────────────────────────────────────────

# Build a reusable prompt (efficient if generating multiple files with the same voice)
print('Building voice clone prompt...')
voice_prompt = model.create_voice_clone_prompt(
    ref_audio=REF_AUDIO,
    ref_text=REF_TEXT,
)

print('Generating audio...')
wavs, sr = model.generate_voice_clone(
    text=TEXT.strip(),
    language=LANGUAGE,
    voice_clone_prompt=voice_prompt,
)

output_path = f"{OUTPUT_DIR}/{SESSION_NAME}.wav"
sf.write(output_path, wavs[0], sr)

print(f'✅ Saved to: {output_path}')
display(Audio(wavs[0], rate=sr))
```

## 5️⃣ Batch generation

Generate multiple sessions in one go and save them all.

```python
import os

# ── Define your batch ─────────────────────────────────────────────────────
SPEAKER  = "Serena"
LANGUAGE = "English"
INSTRUCT = "Speak slowly and gently, with meditative calm and soft pauses between sentences."

sessions = [
    {
        "name": "morning_breath",
        "text": "Good morning. Begin by settling into stillness. Take three deep breaths and feel your body wake up gently with the day."
    },
    {
        "name": "body_scan",
        "text": "Bring your awareness to your feet. Notice any sensation there, without judgment. Slowly travel upward through your legs, your hips, your belly, your chest, your shoulders, your neck, and finally your face. You are whole."
    },
    {
        "name": "loving_kindness",
        "text": "Silently repeat these words. May I be happy. May I be safe. May I be at peace. Now extend these wishes outward, to someone you love, to a stranger, to all beings everywhere."
    },
]
# ──────────────────────────────────────────────────────────────────────────

texts   = [s["text"] for s in sessions]
names   = [s["name"] for s in sessions]

print(f'Generating {len(sessions)} sessions...')
wavs, sr = model.generate_custom_voice(
    text=texts,
    language=[LANGUAGE] * len(sessions),
    speaker=[SPEAKER] * len(sessions),
    instruct=[INSTRUCT] * len(sessions),
)

for wav, name in zip(wavs, names):
    path = os.path.join(OUTPUT_DIR, f"{name}.wav")
    sf.write(path, wav, sr)
    print(f'  ✅ {path}')
    display(Audio(wav, rate=sr))

print('\n🎉 All done!')
```
